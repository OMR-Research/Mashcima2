# Bitmap rendering visual test

This notebook tests renderers that produce raster images from the scene.

```python
import cv2
from IPython.display import Image
from mashcima2.scene.Scene import Scene
from mashcima2.scene.Sprite import Sprite
from mashcima2.scene.ViewBox import ViewBox
from mashcima2.geometry.Rectangle import Rectangle
from mashcima2.geometry.Transform import Transform

dpi = 72
scene = Scene()
view_box = ViewBox(Rectangle(0, 0, 100, 100))

s1 = Sprite.debug_box(scene.space, rectangle=Rectangle(10, 10, 80, 20))
s1.transform = Transform.rotateDegCC(10).then(s1.transform)

s2 = Sprite.debug_box(scene.space, rectangle=Rectangle(10, 40, 80, 20), dpi=5)
s2.transform = Transform.rotateDegCC(10).then(s2.transform)

s3 = Sprite.debug_box(
    scene.space,
    rectangle=Rectangle(40, 40, 80, 20),
    fill_color=(255, 255, 0, 64),
    border_color=(255, 255, 0, 255),
    dpi=150
)
s3.transform = Transform.rotateDegCC(-30).then(s3.transform)

# this shold be culled away
s4 = Sprite.debug_box(scene.space, rectangle=Rectangle(-100, -100, 10, 10))

print(s1.bitmap.shape)
print(s2.bitmap.shape)
print(s3.bitmap.shape)
print(s4.bitmap.shape)
Image(data=cv2.imencode(".png", s2.bitmap)[1], width=600)
```

# OpenCV warp affine bitmap renderer

```python
from mashcima2.rendering.BitmapRenderer import BitmapRenderer
renderer = BitmapRenderer(dpi=dpi)
bitmap = renderer.render(scene, view_box)
Image(data=cv2.imencode(".png", bitmap)[1], width=600)
```
